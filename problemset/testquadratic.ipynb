Import necessary modules.

```python
from quadratic import Quadratic
import numpy as np
import matplotlib.pyplot as plt
```

Construct a quadratic function $f(x) = x^2+2x+1$

```python
f = Quadratic(1, 2, 1)
```

Check class attributes:

```python
print('This quadratic function as quadratic coefficient %.12g, linear coefficient %.12g and constant coefficient %.12g. \
'%(f.a, f.b, f.c)) 
```

Check if value function works correctly:

```python
x = np.array((-1, 0, 1))
ans = np.array((0, 1, 4))
my_val = f.value(x)
ans == my_val
```

Check if plotting function works correctly:

```python
f.plot(-3,1,100)
```
