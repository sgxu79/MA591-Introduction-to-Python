Import necessary modules.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Read the pendulum data.

```python
L = np.arange(0.1, 1.1, 0.1)
T = np.array([0.6, 0.9, 1.1, 1.3, 1.4, 1.6, 1.7, 1.8, 1.9, 2.0])
```

## (a)

```python
plt.plot(T, L, 'ko')
plt.xlabel('T')
plt.ylabel('L')
plt.title('Length of Rod versus Oscillation Period')
```

## (b)

```python
plt.plot(T,L,'ko',label = "Observations")
for i in range(1,4):
    coef = np.polyfit(T, L, i)
    fit = np.poly1d(coef)
    L_fit = fit(T)
    plt.plot(T,L_fit, label = 'Degree ' + str(i) + ' fit')
plt.legend()
plt.title('Polynomial fit of Pendulum data')
```

The linear fit is the worst, while the second and third order polynomial fits are similar.
